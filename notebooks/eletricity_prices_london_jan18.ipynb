```python
df = pd.read_csv("../raw_data/prices_energy_london_jan18.csv")
```

```python
df["timestamp"]=pd.to_datetime(df["Date and time"],format='%H:%M %d-%m-%Y')
del df["Date and time"]
```

```python
df = df[df['Date and time'].dt.month==1]
```

```python
df.set_index('timestamp',inplace=True)
```

```python
df['day'] = df[df['Date and time'].dt.day ]
```

```python
df.resample("H").mean().plot()
```

```python
eletricity_hourly_price = df.resample("H").mean()
```
