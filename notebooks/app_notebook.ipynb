# Visualizations

## Plotly

```python
#
response = requests.get(\
                        'http://localhost:8000/retrofit?building_id=100&meter=0&initial_date=2017-06-01&final_date=2017-06-16')


```

```python
df = response.json()
```

```python
df
```

```python
df = pd.read_json(df)
```

```python
df
```

```python
df.resample("D").mean().plot();
```

```python
df.resample("D").median().plot()
```

```python
daily_consumption = df.resample('D').mean()
```

```python
import plotly.express as px 
fig = px.line(daily_consumption,title=f"Daily Energy Consumption - Building {100}")

fig.show()
```
